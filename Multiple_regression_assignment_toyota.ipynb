# VIF

```python
rsq_age=smf.ols('age~km+hp+cc+doors+gears+qt+wt',data=df).fit().rsquared
vif_age=1/(1-rsq_age)

rsq_km=smf.ols('km~age+hp+cc+doors+gears+qt+wt',data=df).fit().rsquared
vif_km=1/(1-rsq_km)

rsq_hp=smf.ols('hp~age+km+cc+doors+gears+qt+wt',data=df).fit().rsquared
vif_hp=1/(1-rsq_hp)

rsq_cc=smf.ols('cc~age+km+hp+doors+gears+qt+wt',data=df).fit().rsquared
vif_cc=1/(1-rsq_cc)

rsq_doors=smf.ols('doors~age+km+hp+cc+gears+qt+wt',data=df).fit().rsquared
vif_doors=1/(1-rsq_doors)

rsq_gears=smf.ols('gears~age+km+hp+cc+doors+qt+wt',data=df).fit().rsquared
vif_gears=1/(1-rsq_gears)

rsq_qt=smf.ols('qt~age+km+hp+cc+doors+gears+wt',data=df).fit().rsquared
vif_qt=1/(1-rsq_qt)

rsq_wt=smf.ols('wt~age+km+hp+cc+doors+gears+qt',data=df).fit().rsquared
vif_wt=1/(1-rsq_wt)
```

```python
d={'Factors':['age','km','hp','cc','doors','gears','qt','wt'],'vif':[vif_age,vif_km,vif_hp,vif_cc,vif_doors,vif_gears,vif_qt,vif_wt]}
vif_df=pd.DataFrame(d)
vif_df
```

```python
 ##Test for Homoscedasticity or Heteroscedasticity

def get_standardized_value(vals):
    return (vals-vals.mean())/vals.std()
```

# Final Model

```python
while model.rsquared<0.90:
    for c in [np.max(c)>0.5]:
        model=smf.ols('Price~age+km+hp+cc+doors+gears+qt+wt',data=df).fit()
        (c,_)=model.get_influence().cooks_distance
        c
        np.argmax(c),np.max(c)
        df=df.drop(df.index[[np.argmax(c)]],axis=0).reset_index(drop=True)
        df
    else:
        model=smf.ols('Price~age+km+hp+cc+doors+gears+qt+wt',data=df).fit()
        model.rsquared , model.aic
        print('Thus rsquared is improved to',model.rsquared)
```
